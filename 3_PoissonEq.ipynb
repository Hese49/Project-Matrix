## 1. Introduction

The Poisson equation models equilibrium states in physics, such as electrostatic potential and steady heat distribution.

$$
-\frac{d^2 u}{dx^2} = f(x)
$$

✅ Unlike the wave equation, this problem has **no time evolution** — it describes a **static balance**.

👉 We will solve a **linear system instead of simulating dynamics**.


Here are some areas where the Poisson equation is used:
- Electrostatics
- Newtonian Gravity
- Fluid Dynamics
- Heat Conduction
- Image Processing and Computer Graphics
- Computational Fluid Dynamics

---
## 2. Mathematical Model

We consider:

$$
-\frac{d^2 u}{dx^2} = f(x), \quad x \in [0,L]
$$

with boundary conditions:

$$
u(0)=0, \quad u(L)=0
$$

**✅** The solution must **satisfy the equation everywhere at once**.

👉 This is a **global (boundary value) problem**.


---
## 3. Finite Difference Discretization

Using finite differences:

$$
-\frac{u_{i-1}-2u_i+u_{i+1}}{h^2}=f_i
$$

Rewriting:

$$
u_{i-1}-2u_i+u_{i+1}=-h^2 f_i
$$

✅ This is the **same spatial operator** as before.

👉 The **tridiagonal structure** reappears.


---
## 4. Matrix Formulation

We write:

$$A\mathbf{u}=\mathbf{b}$$

where $A$ is tridiagonal matrix.

✅ The PDE becomes a matrix problem.

👉 Solving the PDE = solving a linear system.

```python
#  Build and solve linear system 

import numpy as np
import matplotlib.pyplot as plt

# parameters
L = 1.0
Nx = 100
h = L / Nx

x = np.linspace(0, L, Nx+1)
N = Nx - 1

# build matrix A
A = np.zeros((N, N))
for i in range(N):
    A[i, i] = -2
    if i > 0:
        A[i, i-1] = 1
    if i < N-1:
        A[i, i+1] = 1

# RHS
f = np.sin(np.pi * x[1:-1])
b = -h**2 * f

# solve
u_int = np.linalg.solve(A, b)

# full solution
u = np.zeros(Nx+1)
u[1:-1] = u_int
```

## 6. Visualization of the Solution

✅ The solution reflects the shape of the source term.

👉 The system **smooths the input function**.

```python
#  Plot solution 

plt.figure()
plt.plot(x, u)
plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Solution of Poisson equation')
plt.grid()
plt.show()
```

## 7. Interpretation via Linear Algebra

The matrix represents the discrete Laplacian.

✅ Solving $A\mathbf{u}=\mathbf{b}$ is **operator inversion**.

👉 **Linear algebra defines the problem**.

---
## 8. Connection to Eigenvalues

$$A\mathbf{v}=\lambda\mathbf{v}$$

✅ Same eigenmodes as in the wave equation.

👉 Solution is combination of modes.

```python
#  Eigenmodes 

eigvals, eigvecs = np.linalg.eigh(A)

# Sort eigenvalues (optional but recommended)
idx = np.argsort(eigvals)[::-1]
eigvals = eigvals[idx]
eigvecs = eigvecs[:, idx]


def fix_sign(eigvecs):
    for i in range(eigvecs.shape[1]):
        # Make the second entry positive (or any consistent rule)
        if eigvecs[1, i] < 0:
            eigvecs[:, i] *= -1
    return eigvecs


# Fix zig-zag sign ambiguity
eigvecs = fix_sign(eigvecs)

plt.figure(figsize=(6,4))
for i in range(4):
    v = eigvecs[:, i]
    v_full = np.zeros(Nx+1)
    v_full[1:-1] = v / np.max(np.abs(v))
    plt.plot(x, v_full, label=f'mode {i+1}')

plt.legend()
plt.grid()
plt.title('Eigenmodes')
plt.show()
```

## 9. Summary

$$\text{PDE} \rightarrow \text{matrix equation}$$

✅ **Equilibrium problems** are **linear algebra problems.**

👉 We moved from **simulation** to **direct solution**.
