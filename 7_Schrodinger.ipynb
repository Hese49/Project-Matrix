## 1. Introduction

The Schrödinger equation describes quantum systems.

✅ Unlike previous PDEs, this problem leads naturally to an **eigenvalue problem** instead of a linear system.

👉 **Eigenvalues** correspond to **energy levels**, and **eigenvectors** describe **physical states.**

## 2. Mathematical model

We consider the time-independent Schrödinger equation:
$$
-\frac{d^2 u}{dx^2} = \lambda u
$$
on a domain with boundary conditions.

## 3. Discretization

Using finite differences, we approximate the second derivative.

✅ 
Discretization transforms the differential operator into a matrix whose eigenvalues we compute.

👉 
The continuous operator becomes a matrix eigenvalue problem.

```python
#  Build matrix

import numpy as np

N = 50
h = 1/(N+1)

A = np.zeros((N,N))

for i in range(N):
    A[i,i] = 2
    if i > 0:
        A[i,i-1] = -1
    if i < N-1:
        A[i,i+1] = -1

A = A / h**2
```

```python
#  Solve eigenvalue problem 

eigvals, eigvecs = np.linalg.eigh(A)
```

```python
#  Visualization of first modes 

import matplotlib.pyplot as plt

for k in range(3):
    plt.plot(eigvecs[:,k], label=f"Mode {k}")
plt.legend()
plt.title("First eigenmodes")
plt.show()

print(" Note: The eigenmodes may appear flipped vertically because each eigenvector is defined")
print("  only up to an overall sign, so starting upward or downward has no physical significance.")
```

## 6. Matrix interpretation

The matrix represents the discrete Laplacian.

👉 
**Eigenvectors** correspond to **standing wave modes of the system**.

## 7. Linear algebra insight

Eigenvalues determine system behavior.

✅ 
**Spectral properties** encode **physical information.**

👉 
**Linear algebra** provides direct **access to energy levels and modes.**
