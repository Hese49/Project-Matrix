## 1. Introduction

Large systems from PDEs quickly become expensive if treated as dense matrices.

✅ Most entries in PDE matrices are zero and exploiting this sparsity is essential for efficiency.

👉 Sparse representations drastically **reduce memory and computation cost**.

## 2. Dense vs sparse matrices

A dense matrix stores all entries, while a sparse matrix stores only nonzero values.

👉 
For PDE systems, storing only nonzero entries is enough to fully represent the operator.

```python
#  Dense vs sparse storage

import numpy as np
from scipy.sparse import lil_matrix

N = 50

# dense
A_dense = np.zeros((N*N, N*N))

# sparse
A_sparse = lil_matrix((N*N, N*N))
```

## 3. Sparse representation

We build the same 2D Laplacian matrix using sparse format.

✅ 
The **sparsity pattern** directly reflects **local interactions in the grid**.

```python
#  Build sparse matrix 

def idx(i,j):
    return i*N + j

for i in range(N):
    for j in range(N):
        k = idx(i,j)
        A_sparse[k,k] = 4
        if i > 0:
            A_sparse[k, idx(i-1,j)] = -1
        if i < N-1:
            A_sparse[k, idx(i+1,j)] = -1
        if j > 0:
            A_sparse[k, idx(i,j-1)] = -1
        if j < N-1:
            A_sparse[k, idx(i,j+1)] = -1

A_sparse = A_sparse.tocsr() 
# Converts the matrix into Compressed Sparse Row (CSR) format, which is optimized
# for fast matrix–vector products and efficient solving once the matrix is assembled.

print(A_sparse)
```

## 4. Numerical implementation

We compare solving using dense vs sparse structures.

```python
#  Solve comparison 

from scipy.sparse.linalg import spsolve
import time

b = np.ones(N*N)

# sparse solve
start = time.time()
x_sparse = spsolve(A_sparse, b)
sparse_time = time.time() - start

# dense solve
start = time.time()
x_dense = np.linalg.solve(A_sparse.toarray(), b)
# A_sparse.toarray() converts onverts the sparse matrix into a full dense NumPy array.
dense_time = time.time() - start

print("Sparse time:", sparse_time)
print("Dense time:", dense_time)
```

## 5. Performance comparison

👉 
**Sparse methods scale much better as system size grows.**

## 6. Matrix interpretation

The sparsity pattern corresponds to grid connectivity.

✅ 
**Matrix structure encodes geometry**, not just numbers.

## 7. Linear algebra insight

Sparse matrices enable specialized algorithms.

👉 
**Efficiency** comes from **exploiting structure**, not brute force computation.
