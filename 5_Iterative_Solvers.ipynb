## 1. Introduction

Large linear systems from PDEs can be expensive to solve directly.

✅ Instead of solving $Ax=b$ in one step, iterative methods gradually improve an approximate solution.

👉 Iterative solvers trade exactness for efficiency and scalability.


## 2. Mathematical model

We consider the linear system:
$$
A u = b
$$
coming from PDE discretization.


## 3. Discretization recap

From the 2D Poisson equation, we obtained a sparse matrix system where each point interacts locally.


## 4. Iterative methods

*Jacobi method* updates all values using the previous iteration.

*Gauss–Seidel* uses updated values immediately.

✅ Iterative methods can be seen as repeated applications of a matrix transformation.

👉 Different update strategies lead to different convergence speeds.


```python
#  Setup system 

import numpy as np

N = 20
u = np.zeros((N,N))
f = np.ones((N,N))
h = 1/(N+1)
```

```python
#  Jacobi method 

def jacobi(u, f, iterations=50):
    u_new = u.copy()
    for _ in range(iterations):
        u_old = u_new.copy()
        for i in range(1,N-1):
            for j in range(1,N-1):
                u_new[i,j] = 0.25*(u_old[i+1,j] + u_old[i-1,j] +
                                   u_old[i,j+1] + u_old[i,j-1] -
                                   h**2 * f[i,j])
    return u_new
```

```python
#  Gauss-Seidel method

def gauss_seidel(u, f, iterations=50):
    for _ in range(iterations):
        for i in range(1,N-1):
            for j in range(1,N-1):
                u[i,j] = 0.25*(u[i+1,j] + u[i-1,j] +
                               u[i,j+1] + u[i,j-1] -
                               h**2 * f[i,j])
    return u
```

```python
#  Run methods

u_jacobi = jacobi(np.zeros((N,N)), f)
u_gs = gauss_seidel(np.zeros((N,N)), f)
```

```python
#  Visualization 

import matplotlib.pyplot as plt

plt.imshow(u_jacobi, origin='lower')
plt.title("Jacobi solution")
plt.colorbar()
plt.show()

plt.imshow(u_gs, origin='lower')
plt.title("Gauss-Seidel solution")
plt.colorbar()
plt.show()
```

## 7. Linear algebra insight

Iterative methods correspond to splitting the matrix.

👉 Convergence depends on spectral properties of the iteration matrix.


## 8. Summary

👉 Iterative methods provide scalable tools for solving large systems arising from PDEs.
